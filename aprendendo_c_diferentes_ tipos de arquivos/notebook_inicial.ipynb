```python
import sqlalchemy
from sqlalchemy import create_engine, MetaData, Table, inspect
```

```python
engine = create_engine('sqlite:///:memory:')
```

```python
url = 'clientes_banco.csv'
```

```python
dados = pd.read_csv(url,sep=',')
```

```python
dados
```

```python
dados.to_sql('clientes', engine, index=False)
```

```python
inspector = inspect(engine)
```

```python
print(inspector.get_table_names())
```

```python
query = 'SELECT * FROM clientes WHERE Categoria_de_renda="Empregado"'
```

```python
empregrados = pd.read_sql(query, engine)
```

```python
empregrados
```

```python
empregrados.to_sql('empregrados', con=engine, index=False)
```

```python
pd.read_sql_table('empregrados', engine)
```

```python
pd.read_sql_table('empregrados', engine, columns=['ID_Cliente', 'Rendimento_anual', 'Grau_escolaridade'])
```
