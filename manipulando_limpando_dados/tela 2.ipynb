```python
import numpy as np
import pandas as pd
```

```python
vendas = pd.read_json('dados_vendas_clientes.json')
locacao = pd.read_json('dados_locacao_imoveis.json')
```

```python
vendas = pd.json_normalize(vendas['dados_vendas'])
locacao = pd.json_normalize(locacao['dados_locacao'])
```

```python
vendas = vendas.explode(['Cliente', 'Valor da compra'])
```

```python
vendas.reset_index(inplace=True, drop=True)
```

```python
vendas['Valor da compra'] = vendas['Valor da compra'].map(lambda x: x.replace('R$', '').replace(',','.').strip()).astype(np.float64)
```

```python
vendas.info()
```

```python
col = list(locacao)
```

```python
locacao = locacao.explode(col[1:])
```

```python
locacao.reset_index(inplace=True, drop=True)
```

```python
locacao['valor_aluguel'] = locacao['valor_aluguel'].map(lambda x: x.replace('$', '').replace(',', '.').replace('reais', '').strip())
```

```python
locacao['valor_aluguel'] = locacao['valor_aluguel'].astype(np.float64)
```
